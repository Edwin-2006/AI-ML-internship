```python
# IMPORT LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# LOAD DATASET
df = pd.read_csv("test.csv")

# DISPLAY FIRST 5 ROWS
print(df.head())

# BASIC INFORMATION
print("\nDataset Info:")
print(df.info())

# CHECK NULL VALUES
print("\nMissing Values:")
print(df.isnull().sum())

# DATASET SHAPE
print("\nRows and Columns:")
print(df.shape)

# STATISTICAL SUMMARY
print("\nSummary Statistics:")
print(df.describe())

# -------------------------------
# VISUALIZATION SECTION
# -------------------------------

# HISTOGRAM OF NUMERICAL COLUMNS
df.hist(figsize=(15,10))
plt.tight_layout()
plt.show()

# CORRELATION HEATMAP
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# COUNTPLOT OF CONTINENT
plt.figure(figsize=(8,5))
sns.countplot(x="continent", data=df)
plt.title("Teams by Continent")
plt.xticks(rotation=45)
plt.show()

# FIFA RANK DISTRIBUTION
plt.figure(figsize=(8,5))
sns.histplot(df["fifa_rank_pre_tournament"], bins=10, kde=True)
plt.title("FIFA Rank Distribution")
plt.show()

# GOALS SCORED VS GOALS RECEIVED
plt.figure(figsize=(8,5))
sns.scatterplot(
    x="goals_scored_last_4y",
    y="goals_received_last_4y",
    hue="continent",
    data=df
)
plt.title("Goals Scored vs Goals Received")
plt.show()

# TOP 10 TEAMS BY MARKET VALUE
top_teams = df.sort_values(
    by="squad_total_market_value_eur",
    ascending=False
).head(10)

plt.figure(figsize=(10,6))
sns.barplot(
    x="squad_total_market_value_eur",
    y="team",
    data=top_teams
)
plt.title("Top 10 Teams by Market Value")
plt.show()

# CHECK UNIQUE CONTINENTS
print("\nUnique Continents:")
print(df["continent"].unique())

# CHECK DUPLICATES
print("\nDuplicate Rows:", df.duplicated().sum())

# REMOVE DUPLICATES IF ANY
df = df.drop_duplicates()

# SAVE CLEANED DATA
df.to_csv("cleaned_test.csv", index=False)

print("\nCleaned dataset saved successfully!")
```
